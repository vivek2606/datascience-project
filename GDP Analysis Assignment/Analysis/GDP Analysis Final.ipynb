### Plot the percentage contribution of primary, secondary and tertiary sectors as a percentage of total GDP for all the states.

```python
combined_df.head() # Lets go back to the master dataframe for 1-B analysis.
```

Now Let's calculate the contribution of Primary, Secondary & Tertiary Sectors in total GDP of states.

```python
Primary = combined_df[combined_df['Sub-Sector'] == "Primary"]['2014-15'].sum()
Secondary = combined_df[combined_df['Sub-Sector'] == "Secondary"]['2014-15'].sum()
Tertiary = combined_df[combined_df['Sub-Sector'] == "Tertiary"]['2014-15'].sum()
Total = Primary + Secondary + Tertiary
Primary_Share = (Primary / Total *100).round(1)
Secondary_Share = (Secondary /Total * 100).round(1)
Tertiary_Share = (Tertiary / Total * 100).round(1)
```

```python
(Primary_Share,Secondary_Share,Tertiary_Share) # Percentage Share of Primary, Secondary & Tertiary Sectors as a Tuple. 
```

Let's Plot the contribution as a Pie Chart.

```python
labels = ['Primary','Secondary','Tertiary']
values = [Primary_Share,Secondary_Share,Tertiary_Share]
data = [go.Pie(labels=labels, values=values)]
layout = go.Layout(
    autosize=False,
    width=600,
    height=600,
    title=go.layout.Title(
        text='Contribution of primary, secondary and tertiary sectors in GDP',
        xref='container',
        x=1
    ),)
fig = go.Figure(data=data,layout=layout)

iplot(fig)
```
