# Model Building
Let's start by splitting our data into a training set and a test set.

### Splitting Data into Training and Test Sets

```python
from sklearn.model_selection import train_test_split

# Putting feature variable to X
X = df.drop(['Churn','mobile_number'],axis=1)

# Putting response variable to y
y = df['Churn']

y.head()
```

```python
# Splitting the data into train and test
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7,test_size=0.3,random_state=100)
```

## Standard Scaling

```python
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train[:] = scaler.fit_transform(X_train[:]) 
```

```python
X_test[:] = scaler.transform(X_test[:])
```

## Dimensionality reduction using PCA

```python
X_train.shape
```

```python
X_train.head()
```

```python
X_test.shape
```

```python
X_test.head()
```

```python
#Improting the PCA module
from sklearn.decomposition import PCA
pca = PCA(svd_solver='randomized', random_state=42)
#Doing the PCA on the train data
pca.fit(X_train)
```

```python
pca.components_
```

```python
colnames = list(X_train.columns)
pcs_df = pd.DataFrame({'PC1':pca.components_[0],'PC2':pca.components_[1], 'Feature':colnames})
pcs_df.head()
```

```python
fig = plt.figure(figsize = (15,15))
plt.scatter(pcs_df.PC1, pcs_df.PC2)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
for i, txt in enumerate(pcs_df.Feature):
    plt.annotate(txt, (pcs_df.PC1[i],pcs_df.PC2[i]))
plt.tight_layout()
plt.show()
```

```python
pca.explained_variance_ratio_
```

```python
#Making the screeplot - plotting the cumulative variance against the number of components
fig = plt.figure(figsize = (12,8))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```

We can easily observe that 60 components are good enough to explain 95% variance in the dataset. We will choose No. of components = 60

```python
# Using incremental PCA for efficiency - saves a lot of time on larger datasets
from sklearn.decomposition import IncrementalPCA
pca_final = IncrementalPCA(n_components=60)
```

```python
X_train_pca = pca_final.fit_transform(X_train)
X_train_pca.shape
```

```python
X_test_pca = pca_final.transform(X_test)
```

```python
#creating correlation matrix for the principal components
corrmat = np.corrcoef(X_train_pca.transpose())
```

```python
plt.figure(figsize = (30,30))
sns.heatmap(corrmat,annot = True)
```

```python
# 1s -> 0s in diagonals
corrmat_nodiag = corrmat - np.diagflat(corrmat.diagonal())
print("max corr:",corrmat_nodiag.max(), ", min corr: ", corrmat_nodiag.min(),)
# we see that correlations are indeed very close to 0
```

This shows that there is no correlation among any two components. Hence, we have treated the colliniarity in the dataset.

#### Grid Search to Find Optimal Hyperparameters 

```python
# Create the parameter grid based on the results of random search 
param_grid = {
    'max_depth': [4,8,10],
    'min_samples_leaf': range(100, 400, 200),
    'min_samples_split': range(200, 500, 200),
    'n_estimators': [100,200, 300], 
    'max_features': [10,20,50]
}
# Create a based model
rf = RandomForestClassifier()
# Instantiate the grid search model
grid_search = GridSearchCV(estimator = rf, param_grid = param_grid, 
                          cv = 3, n_jobs = -1,verbose = 1)
```

```python
# Fit the grid search to the data
grid_search.fit(X_train, y_train)
```

```python
# printing the optimal accuracy score and hyperparameters
print('We can get accuracy of',grid_search.best_score_,'using',grid_search.best_params_)
```

**Fitting the final model with the best parameters obtained from grid search.**

```python
# model with the best hyperparameters
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(bootstrap=True,
                             max_depth=8,
                             min_samples_leaf=100, 
                             min_samples_split=200,
                             max_features=50,
                             n_estimators=200)
```

```python
# fit
rfc.fit(X_train,y_train)
```

```python
# evaluation metrics
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
